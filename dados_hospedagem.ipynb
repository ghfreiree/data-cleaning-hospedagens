# Preparação dos dados

```python
import pandas as pd
```

```python
dados = pd.read_json('/content/dados_hospedagem.json')
dados.head()
```

```python
# Normaliza o json para visualizar o dataframe corretamente
dados = pd.json_normalize(dados['info_moveis'])
dados
```

```python
colunas = list(dados.columns)
colunas
```

```python
# Tira os dados dos formatos de lista
dados = dados.explode(colunas[3:])
dados
```

```python
# Reseta os índices do dataframe
dados.reset_index(inplace = True,drop=True)
dados.head()
```

# Manipulação de dados numéricos

```python
dados.info()
```

```python
# Transforma os números máximos de hóspedes para o tipo int
dados['max_hospedes'] = dados['max_hospedes'].astype(int)
```

```python
dados.info()
```

```python
# Colunas com valores numéricos inteiros
col_numericas = ['quantidade_banheiros','quantidade_quartos','quantidade_camas']
```

```python
# Transforma o resto das colunas de valores inteiros para int
dados[col_numericas] = dados[col_numericas].astype(int)
```

```python
# Transforma dados da avaliação em float
dados['avaliacao_geral'] = dados['avaliacao_geral'].astype(float)
```

```python
dados.info()
```

```python
# Limpa os dados de preço para tranformá-los em numéricos
dados['preco'] = dados['preco'].replace(r'[\$,]', '', regex=True)
dados['preco']
```

```python
# Transforma em float
dados['preco'] = dados['preco'].astype(float)
```

```python
dados.info()
```

```python
# Colunas para limpar
colunas_taxas = ['taxa_deposito', 'taxa_limpeza']

dados[colunas_taxas] = dados[colunas_taxas].replace(r'[\$,]', '', regex=True)

# Conversão para float
dados[colunas_taxas] = dados[colunas_taxas].astype(float)

dados[colunas_taxas].head()
```

```python
dados.info()
```

# Tokenização

```python
# Coloca os dados da descrição local todos em minúsculos para a tokenização

dados['descricao_local'] = dados['descricao_local'].str.lower()
```

```python
dados.head()
```

```python
dados['descricao_local'][3169]
```

```python
# Tira os caracteres que não importam das descrições locais
dados['descricao_local'] = dados['descricao_local'].str.replace(r'[^a-zA-Z0-9\-\']',' ',regex=True)
```

```python
dados['descricao_local'][3169]
```

```python
# Tira os hífens isolados das descrições locais
dados['descricao_local'] = dados['descricao_local'].str.replace(r'(?<!\w)-(?!\w)',' ',regex=True)
```

```python
# Splita as descrições para a tokenização
dados['descricao_local'] = dados['descricao_local'].str.split()
dados.head()
```

```python
# Retira as chaves para colocar colchetes
dados['comodidades'] = dados['comodidades'].str.replace(r'\{|}|\"', '', regex=True)
```

```python
# Splita a coluna já com as vírgulas
dados['comodidades'] = dados['comodidades'].str.split(',')
dados.head()
```

# Manipulação de datas

```python
dt_data = pd.read_json('/content/moveis_disponiveis.json')
dt_data
```

```python
dt_data.info()
```

```python
# Transforma as datas para o tipo datetime
dt_data['data'] = pd.to_datetime(dt_data['data'], format='%Y-%m-%d')
```

```python
dt_data.info()
```

```python
dt_data.head()
```

```python
# Tranforma o formato de data para ano-mês
dt_data['data'].dt.strftime('%Y-%m')
```

```python
# Subset com soma das vagas disponíveis por mês
subset = dt_data.groupby(dt_data['data'].dt.strftime('%Y-%m'))['vaga_disponivel'].sum()
subset
```
